```python
#On lit le dossier contenant les données
df = np.load("baseDonnees.npy",allow_pickle=True)
```

```python
#On transforme ces données en tableau de dimension 275246*2
df=list(df)
df=np.array(df)
np.shape(df)
```

```python
#On transforme ce tableau en dataframe et on en renomme les colonnes
import pandas as pd
df2=pd.DataFrame(df)
df2.columns=[ 'img' , 'diag' ]
df2.info()
```

```python
#On trace un diagramme en barres montrant la répartition des diagnostics dans le dataframe
from matplotlib import pyplot as plt
ind=range(1,5)

bar1=len(df2)
bar2=len(df2[df2.diag == 0])
bar3=len(df2[df2.diag == 1])

p1=plt.bar(1,bar1,0.35)
p2=plt.bar(2,bar3,0.35,bottom=bar2)
p25=plt.bar(2,bar2,0.35)
p3=plt.bar(3,bar2,0.35, color="green")
p4=plt.bar(4,bar3,0.35, color="orange")

plt.title("Répartition des diagnostiques")
plt.xticks(ind,('1','2','3','4'))
plt.legend((p1[0],p3[0],p4[0]),('Echantillon total','Positif','Négatif'))
```

```python
#On affiche 10 images
fig = plt.figure(figsize =(10, 10)) 

tabImage=[df2.loc[0,'img'],df2.loc[1,'img'],df2.loc[2,'img'],df2.loc[3,'img'],df2.loc[4,'img'],df2.loc[275235,'img'],df2.loc[275236,'img'],df2.loc[275237,'img'],df2.loc[275238,'img'],df2.loc[275239,'img']]

for i in range(10): 
    plt.subplot(2, 5, i + 1)
    image=tabImage[i].reshape(50,48,3)
    plt.imshow(image) 
    plt.axis('off') 
    plt.savefig('imagesExemple.png')
```
